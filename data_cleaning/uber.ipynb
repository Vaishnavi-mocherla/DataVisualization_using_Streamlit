```python
x = df["Avg CTAT"].mean()
df.fillna({'Avg CTAT': x}, inplace=True)
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
# Select only object (categorical) columns
obj_cols = df.select_dtypes(include="object").columns

# Fill NaN with mode for each object column
for col in obj_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
print(df.duplicated().sum())
```

```python
df.rename(columns=lambda x: x.strip().lower().replace(" ", "_"), inplace=True)
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['time'] = pd.to_datetime(df['time'], errors='coerce')
```

```python
cancellation_by_booking = df.groupby(['cancelled_rides_by_driver', 'vehicle_type']).agg({
    'booking_id' : 'count',
    'booking_value': 'mean',
    'driver_ratings': 'mean',       # number of bookings
    'ride_distance': 'mean'      # avg distance
})
```

```python
cancellation_by_booking
```

```python
incomplete_rides_agg = df.groupby(['incomplete_rides', 'vehicle_type']).agg({
    'booking_id' : 'count',
    'pickup_location' : lambda x: x.mode().iloc[0] if not x.mode().empty else None,
    'ride_distance' : 'mean',



})
```

```python
incomplete_rides_agg
```

```python
customer_perp_agg = df.groupby(['vehicle_type']).agg({
    'customer_rating' : 'mean',
    'avg_vtat' : 'mean',
    'ride_distance' : 'mean',
    'booking_value' : 'mean'


})
customer_perp_agg
```

```python
payments = df.groupby(['payment_method', 'vehicle_type']).agg({
    'ride_distance' :'mean',
    'booking_value' : 'mean',
    'pickup_location' : lambda x: x.mode().iloc[0] if not x.mode().empty else None,
    'drop_location' : lambda x: x.mode().iloc[0] if not x.mode().empty else None

})

payments
```

```python
driver_agg = df.groupby(['driver_cancellation_reason']).agg({
    'pickup_location' : lambda x: x.mode().iloc[0] if not x.mode().empty else None,
    'month' : lambda x: x.mode().iloc[0] if not x.mode().empty else None,
})
driver_agg
```

```python
cus_agg = df.groupby(['reason_for_cancelling_by_customer']).agg({
    'pickup_location' : lambda x: x.mode().iloc[0] if not x.mode().empty else None,
    'month' : lambda x: x.mode().iloc[0] if not x.mode().empty else None,
    'vehicle_type' : lambda x: x.mode().iloc[0] if not x.mode().empty else None,


})
cus_agg
```
